```python
"""If the natural number is greater than 1 and having no positive divisors other than 1 and the number itself etc.
For example: 3, 7, 11 etc are prime numbers.""
#Try to write a program using for loop, if -else to print prime numbers between 0-20.

```

```python
#using for loop and ifelse
start=0
end=20

for i in range(start, end+1):
  if i>1:
    for j in range(2,i):
      if(i % j==0):
        break
    else:
      print(i)
```
